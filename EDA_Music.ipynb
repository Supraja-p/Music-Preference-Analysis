## 3.2 Categorical data analysis

```python
for column in df.loc[:, 'Age':'Frequency [Rock]'].columns:
    lenColor = len(df[column].unique())
    cmap = ['#dcdfe3' for i in range(lenColor)]
    cmap[0] = '#f0d090'
    if np.dtype(df[column]) == 'O':
        fig, ax = plt.subplots(figsize = (10,5))
        ax.bar(df[column].value_counts().index, df[column].value_counts(), color = cmap, alpha = 0.5, edgecolor = 'black', linewidth = 0.8, )
        ax.grid(axis = 'y', alpha = 0.5)
        plt.xticks(rotation=90)
        plt.text(-0.5, df[column].value_counts().max()+50, f'{column} count', fontfamily = 'serif', fontsize = 24, fontweight = 'bold')
        plt.show()
```

# 5. Removing outliers

```python

for column in  df.iloc[:,1:].columns:
    if np.dtype(df[column]) == 'float64':
        q75_q3, q25_q1 = np.percentile(df.loc[:,column], [75,25])
        IQR = q75_q3-q25_q1
        
        upper = q75_q3+(IQR*1.5)
        lower = q25_q1-(IQR*1.5)
    
        
        df.loc[df[column]>upper, column] = np.nan
        df.loc[df[column]<lower, column] = np.nan
        

```

```python

for column in ['Age', 'Hours per day', 'BPM']:
    df[column] = df[column].fillna(np.mean(df[column]))

```
